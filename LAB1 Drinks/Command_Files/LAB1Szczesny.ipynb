```python
import pandas as pd


df = pd.read_csv('../Analysis_Data/drinks.csv')
df = df.drop('total_litres_of_pure_alcohol', axis=1)
df_beer = df[['country', 'beer_servings']]
df_spirit = df[['country', 'spirit_servings']]
df_wine = df[['country', 'wine_servings']]
df_beer.to_csv('../Analysis_Data/drinks_beer.csv', index=False)
df_spirit.to_csv('../Analysis_Data/drinks_spirit.csv', index=False)
df_wine.to_csv('../Analysis_Data/drinks_wine.csv', index=False)

df_test = pd.melt(df, id_vars=['country'], value_vars=['beer_servings','spirit_servings','wine_servings'])
df_test = df_test.sort_values(by='country')
df_test.to_csv('../Analysis_Data/df_tidy.csv', index=False)
df_test
```
